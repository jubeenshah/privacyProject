```python
import yaml
import pandas as pd
import collections
import re
import ast
```

```python
with open("./PrivacyPolicy/apple.yml", 'r') as apple:
    try:
        applePolicyDict = (yaml.safe_load(apple))
    except yaml.YAMLError as exc:
        print(exc)
        
with open("./PrivacyPolicy/google.yml", 'r') as google:
    try:
        googlePolicyDict = (yaml.safe_load(google))
    except yaml.YAMLError as exc:
        print(exc)
        
with open("./PrivacyPolicy/snapchat.yml", 'r') as snapchat:
    try:
        snapchatPolicyDict = (yaml.safe_load(snapchat))
    except yaml.YAMLError as exc:
        print(exc)
        
dictOfAllPolicies = {}
dictOfAllPolicies['Apple'] = applePolicyDict
dictOfAllPolicies['Google'] = googlePolicyDict
dictOfAllPolicies['Snapchat'] = snapchatPolicyDict
```

```python
dash = "-"*50
for index, currentPolicy in enumerate(dictOfAllPolicies):
    print("\n")
    print(dash)
    currentPolicy_list_performed = []
    print("|Current Policy: {:32}|".format(currentPolicy))
    print(dash)
    for index, policy in enumerate(dictOfAllPolicies[currentPolicy]['segments']):
        if policy['annotations'] != []:
        #print(value['annotations'])
            for annotation in policy['annotations']:
                #print(annotation['modality'])
                if annotation['modality'] == 'PERFORMED':
                    #print(annotation)
                    currentPolicy_list_performed.append(annotation['practice'])
    currentPolicy_dict_performed = collections.Counter(currentPolicy_list_performed)
    for index, value in enumerate(currentPolicy_dict_performed):
        print("|{:44s}: {:2s}|".format(str(value), str(currentPolicy_dict_performed[value])))
    print(dash)
```
